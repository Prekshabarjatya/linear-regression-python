# 04 - Model Interpretation

Interpret coefficients and statistical significance from an OLS model.


```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

rng = np.random.default_rng(42)
x = rng.uniform(0, 10, 100)
noise = rng.normal(0, 2, 100)
y = 3 + 2.5 * x + noise
df = pd.DataFrame({"x": x, "y": y})

X = sm.add_constant(df["x"])
model = sm.OLS(df["y"], X).fit()

print(f"Intercept: {model.params['const']:.3f}")
print(f"Slope: {model.params['x']:.3f}")
print(f"R-squared: {model.rsquared:.3f}")
print(f"p-value (slope): {model.pvalues['x']:.3e}")
```
