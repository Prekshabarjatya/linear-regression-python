# 01 - Data Generation for Linear Regression

Create synthetic data with a linear relationship and noise.


```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)
x = rng.uniform(0, 10, 100)
noise = rng.normal(0, 2, 100)
y = 3 + 2.5 * x + noise

df = pd.DataFrame({"x": x, "y": y})
df.head()
```
